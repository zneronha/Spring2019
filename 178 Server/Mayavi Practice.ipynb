```python
from mpl_toolkits import mplot3d
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
zdata = np.linspace(0, 15, 1000)
xdata = np.sin(zdata)
ydata = np.cos(zdata)
ax.plot3D(xdata,ydata,zdata, 'gray')
y1data = xdata
x1data = ydata
```

```python
ax = plt.axes(projection='3d')
ax.plot3D(xdata,ydata,zdata, 'red')
ax.plot3D(x1data,y1data,zdata,'blue')
```
