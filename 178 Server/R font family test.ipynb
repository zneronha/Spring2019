```r
p <- ggplot(mtcars, aes(x=wt, y=mpg)) + geom_point() +
  ggtitle("Fuel Efficiency of 32 Cars") +
  xlab("Weight (x1000 lb)") + ylab("Miles per Gallon") +
  theme(text=element_text(size=16, family="Myriad Pro"))
```

```r
p
```

```r
ggsave("font_ggplot.pdf", plot=p, width=4, height=4)
embed_fonts("font_ggplot.pdf", outfile="font_ggplot.pdf")
```

```r
postscript("font_ggplot.eps", family="Myriad Pro", width=4, height=4, title="Cars Dataset MPG vs. Weight")
p
dev.off()
embed_fonts("font_ggplot.eps", outfile="font_ggplot.eps")
```
